```python
# define function which can print a dictionary where the keys are numbers between 1 and 20  and the values are 
# square of keys

def create_square_dictionary():
    square_dict = {}
    for num in range(1, 21):
        square_dict[num] = num**2
    return square_dict

def print_dictionary(dictionary):
    for key, value in dictionary.items():
        print(f"{key}: {value}")

# Create the dictionary
square_dict = create_square_dictionary()

# Print the dictionary
print_dictionary(square_dict)
```
